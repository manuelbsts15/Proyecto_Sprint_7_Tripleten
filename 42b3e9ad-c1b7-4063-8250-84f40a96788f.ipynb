```python
visit_logs['Start Ts'] = pd.to_datetime(visit_logs['Start Ts'], format="%Y-%m-%d %H:%M")
visit_logs['End Ts'] = pd.to_datetime(visit_logs['End Ts'], format="%Y-%m-%d %H:%M")
```

```python
visit_logs['session_year'] = visit_logs['Start Ts'].dt.year
visit_logs['session_month'] = visit_logs['Start Ts'].dt.month
visit_logs['session_day'] = visit_logs['Start Ts'].dt.day
visit_logs['session_week'] = visit_logs['Start Ts'].dt.isocalendar().week
visit_logs
```

<div class="alert alert-block alert-success">
    <b>Comentarios del Revisor</b> <a class="tocSkip"></a><br>
Muy bien por la extracción de features para este problema, usaste las funciones del objeto .dt adecuadas. </div>

```python
sessions_per_user = visit_logs.groupby(['session_year', 'session_month']).agg({'Uid': ['count','nunique']})
sessions_per_user.columns = ['n_sessions', 'n_users']
sessions_per_user['sess_per_user']  =     sessions_per_user['n_sessions'] / sessions_per_user['n_users']
sessions_per_user
```

```python
# La duración media de la sesión o ASL (Average session length)
visit_logs['session_duration_sec'] = (
    visit_logs['End Ts'] - visit_logs['Start Ts']
).dt.seconds
display(visit_logs['session_duration_sec'].hist(bins=100,
                                              figsize=[5,5])) 
asl=visit_logs['session_duration_sec'].mode()
asl
```

```python
# MAU: el número de usuarios activos mensuales.
mau_total = (
    visit_logs.groupby(['session_year', 'session_month'])
    .agg({'Uid': 'nunique'})
    .mean()
)
print('Numero de usuarios activos mensuales')
mau_total
```

```python
#DAU: el número de usuarios activos diarios (únicos)
dau_total = visit_logs.groupby('session_day').agg({'Uid': 'nunique'}).mean()

print('Numero de usuarios activos diarios')
dau_total
```

```python
# WAU: el número de usuarios activos semanales
wau_total = (
    visit_logs.groupby(['session_year', 'session_week'])
    .agg({'Uid': 'nunique'})
    .mean()
)
print('Numero de usuarios activos semanales')

wau_total
```

```python
#  sticky factor (factor de adherencia)
sticky_wau = dau_total / wau_total * 100
print('Factor de adherencia semanal', sticky_wau)
sticky_mau = dau_total / mau_total * 100
print('Factor de adherencia mensual',sticky_mau)
```

 ## Ventas:
 
¿Cuándo la gente empieza a comprar? (En el análisis de KPI, generalmente nos interesa saber el tiempo que transcurre entre el registro y la conversión, es decir, cuando el/la usuario/a se convierte en cliente. Por ejemplo, si el registro y la primera compra ocurren el mismo día, el/la usuario/a podría caer en la categoría Conversion 0d. Si la primera compra ocurre al día siguiente, será Conversion 1d.  Puedes usar cualquier enfoque que te permita comparar las conversiones de diferentes cohortes para que puedas determinar qué cohorte o canal de marketing es más efectivo).


```python
first_buy = visit_logs.groupby('Uid')['Start Ts'].min().reset_index()
first_buy

ventas = first_buy.merge(order_logs, on='Uid')
ventas['Start Ts'] = ventas['Start Ts'].astype('datetime64[M]')
ventas['Buy Ts'] = ventas['Buy Ts'].astype('datetime64[M]')
ventas['Time'] = ventas['Buy Ts'] - ventas['Start Ts']
ventas
```

```python
ventas.groupby('Time')[['Uid']].count().sort_values(by='Uid', ascending=False).head()
```

Con el resultado obtenido podemos ver que la mayoria de los usuarios prefieren comprar el primer dia que inician sesion.

<div class="alert alert-block alert-success">
    ¡Gran trabajo al identificar la primera venta y al encadenar los métodos de `min`! El uso de `reset_index` es una decisión inteligente para asegurar la consistencia de los índices en futuras operaciones. Además, aplicaste de buena forma la función `merge`.
</div>

## Cuántos pedidos hacen durante un período de tiempo dado?

```python
pedidos_periodo = ventas.groupby('Buy Ts').agg({'Uid':'nunique', 'Revenue':'sum'})  
pedidos_periodo.sort_values(by='Uid', ascending=False)
```

Con este resultado podemos ver que para el mes de diciembre del 2017 se realizaron la mayor cantidad de pedidos.

## ¿Cuál es el tamaño promedio de compra?


```python
pedidos_periodo['Promedio'] = pedidos_periodo['Revenue'] / pedidos_periodo['Uid'] 
display(pedidos_periodo) 
print('El tamaño promdeio de la compra es:', pedidos_periodo['Promedio'].mean())
```

```python
mapa = ventas.pivot_table(
    index='Start Ts',
    columns='Buy Ts',
    values= 'Revenue',
    aggfunc='nunique'
)

plt.figure(figsize=[10,10])
plt.title('Ingresos por cliente')
sns.heatmap(
mapa,
annot=True,
fmt='.0f',
linewidths=1,
linecolor='gray')
```

Con el grafico mostrado podemos observar que los meses junio julio y agosto del 2017 fueron los que mas aporte hubo por parte de los clientes a la empresa.

## Marketing:
 
¿Cuánto dinero se gastó?  (Total/por fuente/a lo largo del tiempo)

```python
order_logs['Buy Ts'] = pd.to_datetime(order_logs['Buy Ts'])
costs['dt'] = pd.to_datetime(costs['dt'])
order_logs['order_month'] = order_logs['Buy Ts'].astype('datetime64[M]')
costs['month'] = costs['dt'].astype('datetime64[M]')
first_orders = order_logs.groupby('Uid').agg({'order_month': 'min'}).reset_index()
first_orders.columns = ['Uid', 'first_order_month']
cohort_sizes = (
    first_orders.groupby('first_order_month')
    .agg({'Uid': 'nunique'})
    .reset_index()
)
cohort_sizes.columns = ['first_order_month', 'n_buyers']
cohort_sizes.head()
```

```python
#asumiendo margen de beneficio 40%
margin_rate = 0.4

orders_ = pd.merge(order_logs, first_orders, on='Uid')
cohorts = (
    orders_.groupby(['first_order_month', 'order_month'])
    .agg({'Revenue': 'sum'}).reset_index())
cohorts
```

<div class="alert alert-block alert-success">
    ¡Gran trabajo! La respuesta es correcta </div>

## ¿Cuánto costó la adquisición de clientes de cada una de las fuentes?

```python
report = pd.merge(cohort_sizes, cohorts, on='first_order_month')

report['gp'] = report['Revenue'] * margin_rate
report['age'] = (
    report['order_month'] - report['first_order_month']
) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')

report['ltv'] = report['gp'] / report['n_buyers']

result = report.pivot_table(
    index='first_order_month', columns='age', values='ltv', aggfunc='mean'
).round()

result = result.fillna('')
result
```

```python
monthly_costs = costs.groupby('month').sum()
monthly_costs
```

## ¿Cuán rentables eran las inversiones? (ROI)

```python
report_ = pd.merge(
    report, monthly_costs, left_on='first_order_month', right_on='month'
)
report_['cac'] = report_['costs'] / report_['n_buyers']

report_['romi'] = report_['ltv'] / report_['cac']
result = report_.pivot_table(
    index='first_order_month', columns='age', values='romi', aggfunc='mean'
)
result = report_.pivot_table(
index='first_order_month',
columns='age',
values='ltv',
aggfunc='mean') 

m6_cum_ltv = result.cumsum(axis=1).mean(axis=0)[5]

print('El LTV promedio durante 6 meses desde el primer pedido:', m6_cum_ltv)
```
